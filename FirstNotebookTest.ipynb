```python
# Import necessary libraries
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Step 1: Define the stock ticker and the number of years for data retrieval
ticker = "AAPL"  # Replace with any ticker you want, e.g., "AAPL" for Apple
years = 5        # Define the number of years to look back

# Step 2: Calculate the start date based on the number of years
end_date = datetime.today()
start_date = end_date - timedelta(days=years * 365)  # Approximate years in days

# Step 3: Download the stock data using yfinance
data = yf.download(ticker, start=start_date, end=end_date)

# Step 4: Plot the stock's closing price
plt.figure(figsize=(10, 5))  # Set the figure size for better visualization
plt.plot(data['Close'], label=f"{ticker} Closing Price", color="blue")

# Add titles and labels for clarity
plt.title(f"{ticker} Stock Price over the Last {years} Years")
plt.xlabel("Date")
plt.ylabel("Price (USD)")

# Show the plot
plt.legend()
plt.show()
```
