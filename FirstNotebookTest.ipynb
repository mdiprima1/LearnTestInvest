```python
# Import necessary libraries
import yfinance as yf
import matplotlib.pyplot as plt

# Step 1: Define the stock ticker and the start and end dates for data retrieval
ticker = "TSLA"               # Define the stock symbol (Tesla in this case)
start_date = "2020-01-01"     # Define the start date for the data (YYYY-MM-DD format)
end_date = "2024-01-01"       # Define the end date for the data (YYYY-MM-DD format)

# Step 2: Download the stock data using yfinance
# The data will include open, high, low, close, volume, and adjusted close prices
data = yf.download(ticker, start=start_date, end=end_date)

# Step 3: Define the SMA period
sma_period = 20               # Define the period for the Simple Moving Average (e.g., 20 days)

# Step 4: Calculate the SMA and add it as a new column in the data
# The rolling() function calculates the rolling mean for the specified window (sma_period)
data['SMA'] = data['Close'].rolling(window=sma_period).mean()

# Step 5: Plot the stock's closing price and SMA
plt.figure(figsize=(10, 5))  # Set the figure size for better visualization

# Plot the closing price
plt.plot(data['Close'], label="Closing Price", color="blue")

# Plot the SMA
plt.plot(data['SMA'], label=f"SMA ({sma_period} days)", color="red")

# Add titles and labels for clarity
plt.title(f"{ticker} Stock Price and {sma_period}-Day SMA")
plt.xlabel("Date")
plt.ylabel("Price (USD)")

# Show the legend to distinguish between closing price and SMA
plt.legend()

# Display the plot
plt.show()
```
